```python
df=df.drop('release_date',axis=1)
df
```

```python
#creating a dataframe with grouped platforms and their respective mean values -- reset_index() is what enables this
platf_scores = df.groupby('platform')['meta_score'].mean().reset_index()

plt.figure(figsize=(50, 10))
bp=plt.bar(x=platf_scores['platform'], height=platf_scores['meta_score'])
plt.xlabel('Platform')
plt.ylabel('Mean Score')
plt.title('Mean Score by Platform')
#to display the values on top of each bar
plt.bar_label(bp, labels=[f"{score:.1f}" for score in platf_scores['meta_score']])

plt.show()
```

```python
year_score = df.groupby('year')['meta_score'].mean().reset_index()
plt.figure(figsize=(50, 10))

plt.plot(year_score['year'], year_score['meta_score'])
plt.xlabel('Year')
plt.ylabel('Mean Score')
plt.title('Mean Score by Year')

plt.show()
```

```python
df['user_review'].dtypes
```

```python
# we can't plot visualization if datatype is object so we will convert it 
df['user_review'] = pd.to_numeric(df['user_review'], errors='coerce').fillna(0)
```

```python
df['user_review'].dtypes
```

```python
year_score = df.groupby('year')['user_review'].mean().reset_index()
plt.figure(figsize=(50, 10))

plt.plot(year_score['year'], year_score['user_review'])
plt.xlabel('Year')
plt.ylabel('Mean user review')
plt.title('Mean user review score by Year')

plt.show()
```
